```python
import torch
```

```python
class BasicModel(torch.nn.Module):
    def __init__(self):
        super().__init__()

        self.layer0 = torch.nn.Linear(in_features=12*128*128, out_features=1024)
        self.layer1 = torch.nn.Linear(in_features=1024, out_features=1024)
        self.layer2 = torch.nn.Linear(in_features=1024, out_features=24*64*64)
    
    def forward(self, features):
        x = torch.tensor(features).view([-1, 12*128*128])

        x = self.layer0(x)
        x = self.layer1(x)
        x = self.layer2(x)

        return x.view([-1, 24, 64, 64]) * 1024.0
```

```python
from dataset import ClimateHackDataset
import xarray as xr
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
```

```python
DATASET_PATH = "gs://public-datasets-eumetsat-solar-forecasting/satellite/EUMETSAT/SEVIRI_RSS/v3/eumetsat_seviri_hrv_uk.zarr"

dataset = xr.open_dataset(
    DATASET_PATH,
    engine='zarr',
    chunks='auto'
)
```

```python
cl_dataset = ClimateHackDataset(dataset)
```

```python
model = BasicModel()
for item in cl_dataset:
    print(model(np.expand_dims(item[1], axis=0)))
    break
```
